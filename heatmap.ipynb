```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
flight_df = sns.load_dataset('flights')

# Pivot the data correctly
flight_df = flight_df.pivot(index="month", columns="year", values="passengers")

# Plot the heatmap
plt.figure(figsize=(12, 6))
ax = sns.heatmap(flight_df, cmap="YlGnBu")  # annot adds values; fmt="d" for integers
plt.title("Monthly Passenger Count (1949-1960)")
plt.ylabel("Month")
plt.xlabel("Year")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.heatmap(flight_df, annot= True, fmt ="d", cmap="YlGnBu")  # annot adds values; fmt="d" for integers
plt.title("Monthly Passenger Count (1949-1960)")
plt.ylabel("Month")
plt.xlabel("Year")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.heatmap(flight_df, annot= True, fmt ="d", linecolor='w',linewidths=5, cmap="YlGnBu",cbar = False)  # annot adds values; fmt="d" for integers
plt.title("Monthly Passenger Count (1949-1960)")
plt.ylabel("Month")
plt.xlabel("Year")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.heatmap(flight_df, annot= True, fmt ="d", linecolor='w',linewidths=5, cmap="YlGnBu")  # annot adds values; fmt="d" for integers
plt.title("Monthly Passenger Count (1949-1960)")
plt.ylabel("Month")
plt.xlabel("Year")
plt.show()
```

```python
grid_kws = {"height_ratios":(.3, 0.05),"hspace":4}
f, (ax,cbar_ax) = plt.subplots(2, gridspec_kw =grid_kws)
ax = sns.heatmap(flight_df, cbar_kws={'orientation':"horizontal"}, ax= ax, cbar_ax =cbar_ax)
```

```python
titanic_df = sns.load_dataset('titanic')
plt.figure(figsize=(12,8))
sns.heatmap(titanic_df.corr())
```

```python
# Encode all categorical variables
encoded_df = pd.get_dummies(titanic_df, drop_first=True)

# Plot heatmap on full encoded data
plt.figure(figsize=(12,8))
sns.heatmap(encoded_df.corr(),  cmap="coolwarm")
plt.title("Correlation Heatmap (Encoded Titanic Data)")
plt.show()
```
