```python
import seaborn as sns
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
sns.barplot(x= 'class', y ='fare', data = titanic_df,hue = 'sex'
                )
```

```python
sns.barplot(x= 'class', y ='fare', data = titanic_df,hue = 'sex', palette= 'icefire'
                )
```

```python
sns.barplot(x= 'class', y ='fare', data = titanic_df,hue = 'sex', palette= 'icefire',
                estimator= np.max)
```

```python
sns.barplot(y= 'class', x='fare', data = titanic_df,hue = 'sex', palette= 'icefire', orient= 'h'
                )
```

```python
sns.barplot(x= 'class', y ='fare', data = titanic_df,hue = 'sex', palette= 'icefire',
                estimator= np.mean)
```

```python
sns.barplot(x= 'class', y ='fare', data = titanic_df,hue = 'sex', palette= 'icefire', ci = 100, errcolor= 'red', errwidth=10
               )
```
