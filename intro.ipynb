```python
bins = [2,3,4,5,6,7,8,9,10]
sns.distplot(df['time_spend_company'], bins = bins)
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins = bins,
            kde = False)
```

```python
sns.distplot(df['time_spend_company'], bins = bins,
            hist = False)
```

```python
sns.distplot(df['time_spend_company'], bins = bins,
            rug = True)
plt.xticks(bins)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram with KDE
sns.histplot(df['time_spend_company'],
             bins=bins,
             kde=True,
             color='red',
             edgecolor='blue',
             linewidth=3,
             alpha=0.5)

# Add rug plot
sns.rugplot(df['time_spend_company'], color='black')

plt.xticks(bins)
plt.show()
```

```python
sns.distplot(df['time_spend_company'], bins = bins,
            rug = True,
            hist_kws= {'color':'red', 'edgecolor':'blue', 'linewidth':'3', 'alpha': 0.5},
            kde_kws= {'color':'orange', 'linewidth':'3', })
```
